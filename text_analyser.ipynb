```python
from tkinter import *
import matplotlib.pyplot as plt 
import string
import re
top = Tk() 
top.geometry('250x275') 
top.title('Text Analyzer')
def zero():
    global enter
    enter=input("Enter text here: \n")
    alpha="a b c d e f g h i j k l m n o p q r s t u v w x y z"

    global alphabets
    alphabets=alpha.split()

    #Frequency of alphabets
    print("\n")
def first():
    print("Frequency of all alphabets in the given text:")
    for i in alphabets:
        if(enter.count(i)>0):
            print(i,": ",enter.count(i))
    
    
def second():  
    #Frequency of words 
    
    print("Frequency of each words")    
    words=enter.split()

    lst_word=list()
    for word in words:
        if(word not in lst_word):
            lst_word.append(word)    
            print(word,": ",words.count(word))
    print("\n")
def third():
    global words
    words=enter.split()
    #Frequency of special characters
    freq_special_char=0
    for word in words:
        for i in word:
            if(i.isalpha()):pass
            elif(i.isdigit()):pass
            else:
                freq_special_char+=1
    print("Frequency of special characters: ",freq_special_char)      
    print("\n")

def fourth():     
    lst=list()
    n=int(input("Enter the number of how many words u want graph visualization: "))
    for i in range(0,n):
        w=input("Enter the word: ")
        lst.append(w)
    freq=list()
    for i in lst:
        x=words.count(i)
        freq.append(x)
    



    plt.bar(lst, freq,width = 0.5) 
    plt.xlabel('words') 
    plt.ylabel('frequency') 
    plt.title('My bar chart!') 
    plt.show()       

    

     

    # Creating Edit box to show selected option 
w = Button(top, text='Enter to enter text',command=zero) 
w.pack() 
w = Button(top, text='Enter to count frequency of alphabets',command=first) 
w.pack() 
w = Button(top, text='Enter to get Frequency of words',command=second) 
w.pack() 
w = Button(top, text='Enter to get frequency of special character',command=third) 
w.pack() 
w = Button(top, text='Enter to get graph',command=fourth) 
w.pack() 
top.mainloop() 
```
